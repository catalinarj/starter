```python
# Importar Matplot

import matplotlib.pyplot as plt
import numpy as np

# Definir variables y valores

days = ("Lunes", "Martes", "Miércoles", "Jueves")
apps = {
    'Tiktok': (112, 53, 28, 108),
    'Instagram': (21, 29, 67, 49),
    'Twitter': (88, 41, 96, 36),
}

# Definir aspectos visuales

x = np.arange(len(days))  # the label locations
width = 0.25  # the width of the bars
multiplier = 0

fig, ax = plt.subplots(layout='constrained')

for attribute, measurement in apps.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    ax.bar_label(rects, padding=3)
    multiplier += 1

# Definir las etiquetas y el título del gráfico
ax.set_ylabel('Minutos')
ax.set_title('Tiempo en redes sociales')
ax.set_xticks(x + width, days)
ax.legend(loc='upper left', ncols=3)
ax.set_ylim(0, 250)

plt.show()
```
